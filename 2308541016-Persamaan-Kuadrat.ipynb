```python
#Membuat code untuk persamaan kuadrat
#input berupa a,b,c berdasarkan rumus umum ax^2+bx+c=0
#1. Mencari solusi menggunakan diskriminan
#2. Menampilkan grafik persamaan kuadrat yang dihasilkan
#3. Menerima input berupa akar-akar yang diketahui dan mencari tahu persamaan yang dihasilkan
#4. Menentukan titik puncak dan sumbu simetri
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

print("Mancari Solusi dari Persamaan Kuadrat Dengan Rumus Umum ax^2+bx+c=0")
print("-------------------------------------------------------------------")

#mengecek nilai a benar tidak sama dengan nol
def validasi_a_16():
    while True:
        a_16 = float(input("Masukan nilai koefisien dari a: "))
        if a_16 != 0:
            return a_16
        print("Nilai a tidak boleh 0. Silakan coba lagi.")

# Menerima input koefisien a, b, c dari user
def menerima_input():
    a_16 = validasi_a_16()
    b_16 = float(input("Masukkan nilai koefisien dari b: "))
    c_16 = float(input("Masukkan nilai koefisien dari c: "))
    return a_16, b_16, c_16

#menghitung diskriminan
def diskriminan_16(a_16, b_16, c_16):
    return b_16**2 - 4*a_16*c_16

#menghitung solusi persamaan kuadrat
def solusi_kuadrat_16(a_16, b_16, c_16):
    D_16 = diskriminan_16(a_16, b_16, c_16)
    x1_16 = (-b_16 + cmath.sqrt(D_16)) / (2*a_16)
    x2_16 = (-b_16 - cmath.sqrt(D_16)) / (2*a_16)
    return x1_16, x2_16, D_16

#kesimpulan dari solusi persamaan kuadrat
def tampilkan_solusi_16(x1_16, x2_16, D_16):
    if x1_16 == x2_16:
        print(f"Solusi dari persamaan kuadrat tersebut adalah x1 = x2 = {x1_16:.2f}")
    else:
        print(f"Solusi dari persamaan kuadrat tersebut adalah x1 = {x1_16:.2f}, dan x2 = {x2_16:.2f}")
    
    if D_16 == 0:
        print("Persamaan kuadrat memiliki akar kembar")
        print(f"Memotong tepat satu titik di sumbu x, yaitu di titik ({x1_16.real:.2f}, 0)")
    elif D_16 < 0:
        print("Persamaan kuadrat memiliki akar imajiner")
        print("Tidak memotong sumbu x")
    else:
        print("Persamaan kuadrat memiliki akar real berbeda")
        print(f"Memotong sumbu x di dua titik, yaitu titik ({x1_16.real:.2f}, 0) dan ({x2_16.real:.2f}, 0)")
        
#menghitung sumbu simetri        
def sumbu_simetri_16 (b_16, a_16):
    return -b_16/2*a_16

#membuat grafik
def grafik_hasil_16(a_16, b_16, c_16, x1_16, x2_16):
    D_16 = diskriminan_16(a_16, b_16, c_16)
    tpx_16 = -b_16 / (2*a_16)
    tpy_16 = -D_16 / (4*a_16)
    print(f"Koordinat titik puncak adalah ({tpx_16:.2f}, {tpy_16:.2f})")
    if a_16 < 0:
        print("Grafik akan terbuka ke bawah")
    else:
        print("Grafik akan terbuka ke atas")
    
    # Menentukan rentang nilai x
    x_16 = np.linspace(tpx_16 - 10, tpx_16 + 10, 400)
    y_16 = a_16*x_16**2 + b_16*x_16 + c_16

    plt.figure(figsize=(8, 8))
    plt.plot(x_16, y_16, label=f'{a_16}x^2 + {b_16}x + {c_16} = 0', color='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Kuadrat')
    plt.show()

#mencari tahu titik puncak persamaan kuadrat berada di kuadran berapa
def posisi_tp_16(a_16, b_16, c_16, x1_16, x2_16):
    D_16 = diskriminan_16(a_16, b_16, c_16)
    tpx_16 = -b_16 / (2*a_16)
    tpy_16 = -D_16 / (4*a_16)
    
    if tpx_16 <= 0 and tpy_16 <= 0 :
        print ("Posisi titik puncak berada di kuadran III")
    elif tpx_16 >= 0  and tpy_16 >= 0 :
        print ("Posisi titik puncak berada di kuadran I")
    elif tpx_16 >= 0 and tpy_16 <= 0 :
        print ("Posisi titik puncak berada di kuadran IV")
    elif tpx_16 <= 0 and tpy_16 >= 0 :
        print ("Posisi titik puncak  berada di kuadran II")
    else :
        print ("Posisi titik puncak berada di tepat di titik pusat koordinat")


#menampilkan persamaan kuadrat bersadarkan masukan user
a_16, b_16, c_16 = menerima_input()
print(f"Persamaan yang kamu masukkan adalah {a_16}x^2 + {b_16}x + {c_16} = 0")

#menghitung dan menampilkan solusi dari persamaan kuadrat
x1_16, x2_16, D_16 = solusi_kuadrat_16(a_16, b_16, c_16)
tampilkan_solusi_16(x1_16, x2_16, D_16)

#menghitung dan menampilkan sumbu simetri
simetri_16 = sumbu_simetri_16(b_16, a_16)
print(f"Sumbu simetri pada persamaan kuadrat ini adalah x = {simetri_16}")

#menampilkan letak titik puncak
posisi_tp_16(a_16, b_16, c_16, x1_16, x2_16)

#menampilkan grafik dari persamaan yang diinput user
grafik_hasil_16(a_16, b_16, c_16, x1_16, x2_16)
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("Mencari Persamaan Kuadrat Dengan Rumus Umum ax^2 + bx + c = 0 Apabila Diketahui Akar-Akarnya")
print("_______________________________________________________________________________________")

# Menerima input akar-akar dari persamaan kuadrat yang diketahui
def input_x_16():
    x1_16 = float(input("Masukan nilai akar persamaan kuadrat pertama yang diketahui: "))
    x2_16 = float(input("Masukkan nilai akar persamaan kuadrat kedua yang diketahui: "))
    return x1_16, x2_16

#menghitung nilai koefisien b
def koefb_16(x1_16, x2_16):
    return -1 * (x1_16 + x2_16)

#menghitung nilai koefisien c
def koefc_16(x1_16, x2_16):
    return x1_16 * x2_16

#menghitung diskriminan
def diskriminan_16(b_16, c_16):
    return b_16**2 - 4 * c_16

#mencari tahu ada tidaknya titik potong
def solusi_16(x1_16, x2_16, D_16):
    if D_16 == 0:
        print(f"Memotong tepat satu titik di sumbu x, yaitu di titik ({x1_16:.2f}, 0)")
    elif D_16 < 0:
        print("Tidak memotong sumbu x")
    else:
        print(f"Memotong sumbu x di dua titik, yaitu titik ({x1_16:.2f}, 0) dan ({x2_16:.2f}, 0)")

#menghitung sumbu simetri
def sumbu_simetri_16(b_16):
    return -b_16 / 2

#membuat grafik
def grafik_hasil_16(b_16, c_16, x1_16, x2_16):
    D_16 = diskriminan_16(b_16, c_16)
    tpx_16 = -b_16 / 2
    tpy_16 = -D_16 / 4
    print(f"Koordinat titik puncak adalah ({tpx_16:.2f}, {tpy_16:.2f})")
    
    # Menentukan rentang nilai x
    x_16 = np.linspace(tpx_16 - 10, tpx_16 + 10, 400)
    y_16 = x_16**2 + b_16 * x_16 + c_16

    plt.figure(figsize=(8, 8))
    plt.plot(x_16, y_16, label=f'x^2 + {b_16}x + {c_16} = 0', color='green')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Kuadrat')
    plt.show()

#mencari tahu titik puncak persamaan kuadrat berada di kuadran berapa
def posisi_tp_16(b_16, c_16, x1_16, x2_16):
    D_16 = diskriminan_16(a_16, b_16, c_16)
    tpx_16 = -b_16 / (2*a_16)
    tpy_16 = -D_16 / (4*a_16)

    if tpx_16 <= 0 and tpy_16 <= 0 :
        print ("Posisi titik puncak berada di kuadran III")
    elif tpx_16 >= 0  and tpy_16 >= 0 :
        print ("Posisi titik puncak berada di kuadran I")
    elif tpx_16 >= 0 and tpy_16 <= 0 :
        print ("Posisi titik puncak berada di kuadran IV")
    elif tpx_16 <= 0 and tpy_16 >= 0 :
        print ("Posisi titik puncak  berada di kuadran II")
    else :
        print ("Posisi titik puncak berada di tepat di titik pusat koordinat")
        
# Meminta input dari pengguna
x1_16, x2_16 = input_x_16()

# Menghitung koefisien b dan c
b_16 = koefb_16(x1_16, x2_16)
c_16 = koefc_16(x1_16, x2_16)

# Menampilkan persamaan kuadrat
print(f"Persamaan yang kamu miliki adalah x^2 + {b_16}x + {c_16} = 0")

# Menghitung dan menampilkan diskriminan
D_16 = diskriminan_16(b_16, c_16)
print(f"Diskriminan dari persamaan ini adalah {D_16}")

# Menampilkan solusi persamaan kuadrat
solusi_16(x1_16, x2_16, D_16)

# Menghitung dan menampilkan sumbu simetri
simetri_16 = sumbu_simetri_16(b_16)
print(f"Sumbu simetri pada persamaan ini adalah x = {simetri_16}")

# Menampilkan letak titik puncak persamaan kuadrat
posisi_tp_16(b_16, c_16, x1_16, x2_16)

# Menampilkan grafik hasil
grafik_hasil_16(b_16, c_16, x1_16, x2_16)
```
