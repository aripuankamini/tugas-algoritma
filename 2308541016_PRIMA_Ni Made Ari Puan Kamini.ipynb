```python
#tugas menentukan bilangan prima dari range yang telah ditentukan (1000n, 1000(n+1))
import math

print("Bilangan Prima Yang Berada Dianatara 16000 - 17000 sebagai berikut:")
print(" ")
akar_16 = math.sqrt(17000) #menghitung akar dari batas atas yang akan menjadi batas atas yang baru di range
pembulatanakar_16 = math.floor(akar_16) #membulatkan kebawah hasil akar batas atas
prima_16 = [] #list untuk menampung bilangan prima

# memberikan nilai bil_16 1000n (n = 2 nim terakhir = 16)
bil_16 = 16000

# Loop untuk memeriksa bilangan dari 16000 hingga 1699 (batas atas = 1000(n+1)= 1000(17))
while bil_16 < 17000:
    # diasumsikan bahwa bilangan yang dicek adalah prima
    cek_prima_16 = True
    # Cek apakah bilangan tersebut memiliki pembagi selain 1 dan dirinya sendiri
    for i_16 in range(2, pembulatanakar_16 + 1):
        if bil_16 % i_16 == 0: #jika hasil modulo 0 berarti ada pembagia selain 1 dan dirinya sendiri
            cek_prima_16 = False #asumsi awal harus diingkar
            break
    # Jika bilangan tersebut prima, tambahkan ke list
    if cek_prima_16:
        prima_16.append(bil_16)
    # Cek bilangan berikutnya
    bil_16 += 1

print(prima_16)
```
