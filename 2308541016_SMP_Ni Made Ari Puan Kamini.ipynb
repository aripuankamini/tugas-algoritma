```python
import cmath

# Menghitung x sebagai bilangan kompleks
x = (-1 - cmath.sqrt(5)) / 2
print(f"x = {x}")

# Menghitung pembilang
pembilang = cmath.log(x**2024,10) - cmath.log(x**2025,10)
print(f"pembilang = {pembilang.real}")

# Menghitung penyebut
penyebut = cmath.log(x * (1 - x), 10)
print(f"penyebut = {penyebut.real}")

# Menghitung solusi
solusi = pembilang / penyebut
print(f"solusi = {solusi.real}")
```

```python
#akar positif
import cmath

# Menghitung x sebagai bilangan kompleks
x_16 = (-1 + cmath.sqrt(5)) / 2
print(f"x = {x_16.real}")

# Menghitung jawaban dari persaman dengan menyederhanakan eksponensial
# log(a^b) = b * log(a)
#log(a)/log(b) = log_b(a) sehingga penyebut dalam fungsi ini dapat dijadikan sebagai basis

# Basis yang diinginkan untuk logaritma
basis_16 = x_16*(1-x_16)  

# Menghitung logaritma dengan basis tertentu
def log_b16(z_16, b_16):
    return cmath.log(z_16) / cmath.log(b_16)

jawab_16 = 2024 * log_b16(x_16, basis_16) + log_b16(1-x_16, basis_16)
solusi_16 = jawab_16.real
print(f"solusi nilai n = { solusi_16:.2f}")
```

```python
#akar negatif
import cmath

# Menghitung x sebagai bilangan kompleks
x_16 = (-1 - cmath.sqrt(5)) / 2
print(f"x = {x_16.real}") #.real digunakan agar hanya bilangan real yang ditampilkan

# Menghitung jawaban dari persaman dengan menyederhanakan eksponensial
# log(a^b) = b * log(a)
#log(a)/log(b) = log_b(a) sehingga penyebut dalam fungsi ini dapat dijadikan sebagai basis

# Basis yang diinginkan untuk logaritma
basis_16 = x_16*(1-x_16) 

# Menghitung logaritma dengan basis tertentu
def log_b16(z_16, b_16):
    return cmath.log(abs(z_16)) / cmath.log(abs(b_16)) #penambahan abs (absolute) disini bertujuan agar mengubah nilai menjadi positif


#nilai x^2024 - x^2025 = x^2024(1-x)
#Pada logaritma log (ab)= log a + log b
jawab_16 = 2024 * log_b16(x_16, basis_16) + log_b16(1-x_16, basis_16)
solusi_16 = jawab_16.real
print(f"solusi nilai n = { solusi_16:.2f}") #pembatasan input hanya 2 desimal
```
