```python
#2.19 Fungsi dalam Phyton

import numpy as np

apk_16 = print("Pada kesempatan kali ini kita akan menghitung luas ataupun volume bangun ruang!")
apk_16 = print("Perlu diperhatikan, program tidak bisa menghitung semua jenis bangun ruang ya ^^")
apk_16 = print ("Berikut pilihan bangun ruang yang bisa program bantu untuk hitung!")
apk_16 = print ("Kubus, Balok, Tabung, Bola")
apk_16 = print (" ")
def Lkubus_16 (R_16):
    lk_16 = 6*(R_16**2)
    return lk_16
def Vkubus_16 (R_16):
    vk_16 = R_16**3
    return vk_16
def Lbalok_16 (p_16,l_16,t_16):
    lbk_16 = 2*(p_16*l_16 + p_16*t_16 + l_16*t_16)
    return lbk_16
def Vbalok_16 (p_16,l_16,t_16):
    vblk_16 = p_16*l_16*t_16
    return vblk_16
def Lstabung_16 (r_16, t_16):
    lst_16 = 2*np.pi*(r_16*t_16)
    return lst_16
def Ltabung_16 (r_16, t_16):
    lt_16 = 2*np.pi*r_16*t_16 + 2*np.pi*(r_16**2)
    return lt_16
def Vtabung_16 (r_16, t_16):
    vt_16 = np.pi*(r_16**2)*t_16
    return vt_16
def Lbola_16 (r_16):
    lbl_16 = 4*np.pi*(r_16**2)
    return lbl_16
def Vbola_16 (r_16):
    vbl_16 = np.pi*(r_16**3)*3/4
    return vbl_16

selesai_16 = True
while  selesai_16 :
    bangun_16 = input("Masukkan jenis bangun ruang yang ingin kamu ketahui : ")
    satuan_16 = input("Masukkan satuan panjang dari bangun ruang tersebut : ")
    if bangun_16.lower() == "kubus":
        volas_16 = input("Kamu ingin menghitung luas atau volume?")
        if volas_16.lower() == "luas":
            R_16 = float(input("Masukkan panjang rusuk kubus yang diketahui : "))
            luas_kubus_16 = Lkubus_16(R_16)
            print ("Luas kubus dengan panjang rusuk", R_16, "adalah", luas_kubus_16, satuan_16, "^2")
        elif volas_16.lower() =="volume":
            R_16 = float(input("Masukkan panjang rusuk kubus yang diketahui : "))
            vol_kubus_16 = Vkubus_16(R_16)
            print ("Volume kubus dengan panjang rusuk", R_16, "adalah", vol_kubus_16, satuan_16, "^3")
        else :
            print ("Mohon maaf masukkanmu tidak dapat kami proses!")
    elif bangun_16.lower() == "balok":
        volas_16 = input("kamu ingin menghitung luas atau volume?")
        if volas_16.lower() == "luas":
            p_16 = float(input("Masukkan panjang balok yang diketahui : "))
            l_16 = float(input("Masukkan lebar balok yang diketahui : "))
            t_16 = float(input("Masukkan tinggi balok yang diketahui : "))
            luas_balok_16 = Lbalok_16(p_16,l_16,t_16)
            print ("Luas balok dengan panjang", p_16, ", lebar ", l_16, ", dan tinggi ", t_16, " adalah ", luas_balok_16, satuan_16, "^2")
        elif volas_16.lower() == "volume":
            p_16 = float(input("Masukkan panjang balok yang diketahui : "))
            l_16 = float(input("Masukkan lebar balok yang diketahui : "))
            t_16 = float(input("Masukkan tinggi balok yang diketahui : "))
            vol_balok_16 = Vbalok_16(p_16,l_16,t_16)
            print ("Luas balok dengan panjang", p_16, ", lebar ", l_16, ", dan tinggi ", t_16, " adalah ", vol_balok_16, satuan_16, "^3")
        else :
            print ("Mohon maaf masukkanmu tidak dapat kami proses!")
    elif bangun_16.lower() == "tabung":
        volas_16 = input("kamu ingin menghitung luas atau volume? ")
        if volas_16.lower() == "luas":
            r_16 = float(input("Masukkan panjang rusuk tabung yang diketahui : "))
            t_16 = float(input("Masukkan tinggi tabung yang diketahui : "))
            luas_selimut_16 = Lstabung_16(r_16,t_16)
            luas_tabung_16 = Ltabung_16(r_16,t_16)
            print ("Luas selimut tabung dengan jari-jari ", r_16, "dan tinggi ", t_16, "adalah", luas_selimut_16, satuan_16, "^2" "dan luas permukaan tabung adalah ", luas_tabung_16, satuan_16, "^2")
        elif volas_16.lower() == "volume":
            r_16 = float(input("Masukkan panjang rusuk tabung yang diketahui : "))
            t_16 = float(input("Masukkan tinggi tabung yang diketahui : "))
            vol_tabung_16 = Vtabung_16(r_16,t_16)
            print ("Volume dari tabung dengan jari-jari sebesar ", r_16, " dan tinggi ", t_16, "adalah", vol_tabung_16, satuan_16, "^2")
        else:
            print("Mohon maaf masukkanmu tidak dapat kami proses!")
    elif bangun_16.lower() == "bola":
        volas_16 = input("kamu ingin menghitung luas atau volume? ")
        if volas_16.lower() == "luas":
            r_16 = float(input("Masukkan panjang rusuk yang diketahu : "))
            luas_bola_16 = Lbola_16(r_16)
            print ("Luas bola dengan jari-jari ", r_16, "adalah ", luas_bola_16, satuan_16, "^2")
        elif volas_16.lower() == "volume":
            r_16 = float(input("Masukkan panjang rusuk yang diketahu : "))
            vol_bola_16 = Vbola_16
            print ( "Volume bola dengan jari-jari ", r_16, " adalah", vol_bola_16, satuan_16, "^3")
        else:
            print("Mohon maaf masukkanmu tidak dapat kami proses!")
    else:
        print("Mohon maaf bangun ruang ", bangun_16, "tidak terdapat dalam pilihan.")
        print("Pilihan yang program sediakan hanya : KUBUS, BALOK, TABUNG, BOLA")
        
    lanjut_16 = str(input("lagi? ya/tidak: "))
    if lanjut_16.lower() == "tidak":
        selesai_16 = False
        
print ("Terima Kasih <3")
```
