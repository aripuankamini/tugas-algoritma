```python
import numpy as np
apk_16 = print("Kali ini kami akan mencari SUM, MIN, MAX hingga MEAN")
apk_16 = print("__________________________________________________________")
apk_16 = print(" ")


selesai_16 = True
while selesai_16 :
    matriks_16 = input("Masukkan sederet bilangan yang akan dijadikan matriks. Bilangan dipisah menggunakan spasi " ".")
    ubah_list_16 = matriks_16.split()
    data_16 = np.array([float(x_16) for x_16 in ubah_list_16])
    print ("Data yang anda masukkan adalah", data_16)
    end_16 = True
    while end_16:
        operator_16 = input("Apa yang ingin anda cari dari data tersebut : ")
        if operator_16.lower() == "sum":
            sum_16 = np.sum(data_16)
            print("Total penjumlahan dari data yang anda masukkan adalah", sum_16)
        elif operator_16.lower()== "min":
            min_16 = np.min(data_16)
            print("Nilai minimum dari data yang anda masukkan adalah", min_16)
        elif operator_16.lower()== "max":
            max_16 = np.max(data_16)
            print("Nilai maksimum dari data yang anda masukkan adalah", max_16)
        elif operator_16.lower() == "mean":
            mean_16 = np.mean(data_16)
            print("Nilai rata-rata dari data yang anda masukkan adalah", mean_16)
        else: 
            print("Maaf masukkan anda salah")
        go_16 = str(input("Apakah ada yang ingin dicari lagi? y/n "))
        if go_16.lower() == "n":
            end_16 = False
    lanjut_16 = str(input("lagi? y/n"))
    if lanjut_16.lower() == "n":
        selesai_16 = False
print ("terima kasih")
```
