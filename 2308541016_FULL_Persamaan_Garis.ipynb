```python
#persamaan garis yang melewati 2 titik
#terdapat 3 kemungkinan dari kedua persamaan garis:
#kemungkinan 1: terdapat perpotongan di satu titik (gradien kedua garis tidak sama)
#kemungkinan 2: terdapat tidak berhingga banyaknya solusi ( kedua garis berhimpitan)
#kemungkinan 3: tidak terdapat perpotongan garis (tidak memiliki solusi)

#program akan dibuat untuk mencari titik potong dari dua garis yang memenuhi salah satu dari ketiga kasus dibawah ini:
#kasus 1: garis melewati 2 titik
#kasus 2: garis diketahui gardien dan titik yang dilewati
#kasus 3: mengikuti persamaan garis ax+by=c dan dx+ey=f
```

```python
import numpy as np
import matplotlib.pyplot as plt

apk_16 = print("Program akan mencari tahu titik potong dari dua buah persamaan garis berdasarkan 3 kondisi seperti di bawah ini")
apk_16 = print("----------------------------------------------------------------------------------------------------------------")

mulai_16 = True
while mulai_16 :
    apk_16 = print(" [1] : Garis melewati 2 titik")
    apk_16 = print(" [2] : Garis diketahui gradien dan titik yang dilewati")
    apk_16 = print(" [3] : Mengikuti persamaan garis ax+by=c dan dx+ey=f")
    apk_16 = print(" [4] : Keluar")
    
    apk_16 = print(" ")
    
    pilihan_16 = input("Kondisi yang ingin anda cari solusinya : ")
    
    if pilihan_16 == "1":
        print(" Kondisi yang ingin kamu cari solusinya adalah garis melewati 2 titik")
        print(" ")
        #titik persamaan garis yang pertama
        titikx1_16 = float(input("Masukkan titik x pertama: ")) #input koordinat (x1,y1)
        titiky1_16 = float(input("Masukkan titik y pertama: "))
        titikx2_16 = float(input("Masukkan titik x kedua : "))  #input koordinat (y1,y2)
        titiky2_16 = float(input("Masukkan titik y kedua : "))
        print(" ")
        #titik persamaan garis yang kedua
        titikx3_16 = float(input("Masukkan titik x pertama: "))  #input koordinat (x3,y3)
        titiky3_16 = float(input("Masukkan titik y pertama: "))
        titikx4_16 = float(input("Masukkan titik x kedua : "))   #input koordinat (x4,y4)
        titiky4_16 = float(input("Masukkan titik y kedua : "))
        print(" ")
        #menampilkan titik-titik yang telah diinput sebelumnya
        apk_16 = print(f'Titik pertama ({titikx1_16}, {titiky1_16}), dan titik kedua ({titikx2_16}, {titiky2_16})')
        apk_16 = print(f'Titik ketiga ({titikx3_16}, {titiky3_16}), dan titik keempat ({titikx4_16}, {titiky4_16})')

        #mencari gradien
        m1_16 = (titikx1_16 - titikx2_16)/(titiky1_16 - titiky2_16)
        print(f'Gradien dari kedua titik pertama  adalah {m1_16}')  #menampilkan gradien 1
        persamaan1_16 = print(f'Persamaan garis pertama y = {m1_16}x + {(m1_16*titikx1_16)*-1 + titiky1_16}') #menampilkan persamaan garis 1
        m2_16 = (titikx3_16 - titikx4_16)/(titiky3_16 - titiky4_16)
        print(f'Gradien dari kedua titik terakhir adalah {m2_16}') #menampilkan gradien 2
        persamaan2_16 = print(f'Persamaan garis kedua y = {m2_16}x + {(m2_16*titikx3_16)*-1 + titiky3_16}') #menampilkan persamaan garis 2

        #mendefinisikan variabel c pada persamaan 1 dan 2
        c1_16 = -m1_16*titikx1_16 + titiky1_16
        c2_16 = -m2_16*titikx3_16 + titiky3_16
        #mengevaluasi gradien untuk mencari tahu apakah memiliki solusi atau tidak
        if m1_16 == m2_16 : 
            print ("Gradiennya sama nih! Jadi kemungkinan tidak ada solusi (sejajar) atau solusinya sangat banyak (berhimpit)!")
            if c1_16 == c2_16: #mencari tahu apakah kedua garis tersebut berhimpit atau sejajar dengan melihat nilai dari variabel c
                print("Wahhh Kedua garis berhimpit! Jadi Persamaan Tersebut Memiliki Banyak Solusi")
            else : 
                print("Tampaknya Kedua Garismu Sejajar! Yahh, Persamaanmu Tidak Memiliki Solusi nih")
            x_16 = np.arange(c1_16-5,c1_16+5,0.01)
            plt.figure(figsize=(8,8),dpi=200)
            plt.plot(x_16, m1_16*(x_16 - titikx1_16)+titiky1_16, color = 'pink', label = 'y = {}x+{}'.format(m1_16,(m1_16*titikx1_16)*-1 + titiky1_16 ))
            plt.plot(x_16, m2_16*(x_16 - titikx3_16)+titiky3_16, color = 'blue', label = 'y = {}x+{}'.format(m2_16,(m2_16*titikx3_16)*-1 + titiky3_16 ))
            plt.axhline(0, color='black', linewidth=0.5)
            plt.axvline(0, color='black', linewidth=0.5)
            plt.grid(color='gray', linestyle='--', linewidth=0.5)
            plt.title('Grafik 2 Fungsi Linear', fontdict={'fontname':'Times New Roman', 'fontsize': 14}) #mengganti font yang digunakan pada judul grafik
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.legend(loc='best')
            plt.show()
        else : #karena m1 =/= m2 sehingga memiliki solusi tunggal
            titikpotongx_16 = (m1_16*titikx1_16-titiky1_16-m2_16*titikx3_16+titiky3_16)/(m1_16 - m2_16) #mencari titik potong x
            titikpotongy_16 = m2_16*(titikpotongx_16-titikx3_16) + titiky3_16   #mencari titik potong y
            print(f'Koordinat titik potong dari dua persamaan tersebut adalah ({titikpotongx_16}, {titikpotongy_16})') #menampilkan koorddinat titik potong

            x_16 = np.arange(titikpotongx_16-5,titikpotongx_16+5,0.01)
            plt.figure(figsize=(4,4),dpi=200)
            plt.plot(x_16, m1_16*(x_16 - titikx1_16)+titiky1_16, color = 'pink', label = 'y = {}x+{}'.format(m1_16,(m1_16*titikx1_16)*-1 + titiky1_16 ))
            plt.plot(x_16, m2_16*(x_16 - titikx3_16)+titiky3_16, color = 'blue', label = 'y = {}x+{}'.format(m2_16,(m2_16*titikx3_16)*-1 + titiky3_16 ))
            plt.plot(titikpotongx_16,titikpotongy_16, color='orange', marker='*',markersize=6) #memberi tanda pada titik potong
            plt.axhline(0, color='black', linewidth=0.5)
            plt.axvline(0, color='black', linewidth=0.5)
            plt.grid(color='gray', linestyle='--', linewidth=0.5)
            plt.title('Grafik 2 Fungsi Linear', fontdict={'fontname':'Times New Roman', 'fontsize': 14}) #mengganti font yang digunakan pada judul grafik
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.legend(loc='best')
            plt.show()
    elif pilihan_16 == "2":
        print(" Kondisi yang ingin kamu cari solusinya adalah garis diketahui gradien dan titik yang dilewati")
        apk_16 = print("Gradien yang dapat kami proses hanya dalam bentuk bilangan bulat dan pecahan (p/q).")
        ask_16 = input("Apakah gradien yang diketahui dalam bentuk bilangan bulat? (y/n)") #untuk memastikan input dari user adalah integer

        if ask_16.lower() == "y":
            m1_16 = float(input("Masukkan gradien pertama yang diketahui (untuk desimal gunakan titik '.'): "))
            m2_16 = float(input("Masukkan gradien kedua yang diketahui (untuk desimal gunakan titik '.'): "))
            tx1_16 = float(input("Masukkan koordinat titik x pertama : "))
            ty1_16 = float(input('Masukkan koordinat titik y pertama : '))
            tx2_16 = float(input("Masukkan koordinat titik x kedua : "))
            ty2_16 = float(input('Masukkan koordinat titik y kedua : '))
            #menampilkan gradien, koordinat, dan persamaan garis yang telah diinput user
            print("persamaan garis dengan gradien", m1_16, "dan pada koordinat titik (",{tx1_16},",",{ty1_16},") adalah y = {:.2f}x+{:.2f}".format(m1_16, -m1_16*tx1_16+ty1_16))
            print("persamaan garis dengan gradien", m2_16, "dan pada koordinat titik (",{tx2_16},",",{ty2_16},") adalah y = {:.2f}x+{:.2f}".format(m2_16, -m2_16*tx2_16+ty2_16))
            
            c1_16 = (m1_16*tx1_16)*-1 + ty1_16
            c2_16 = (m2_16*tx2_16)*-1 + ty2_16
            #mengevaluasi nilai gradien
            if m1_16 == m2_16 : 
                print ("Gradiennya sama nih! Jadi kemungkinan tidak ada solusi (sejajar) atau solusinya sangat banyak (berhimpit)!")
                if c1_16 == c2_16: #mencari tahu apakah kedua garis tersebut berhimpit atau sejajar dengan melihat nilai dari variabel c
                    print("Wahhh Kedua garis berhimpit! Jadi Persamaan Tersebut Memiliki Banyak Solusi")
                else : 
                    print("Tampaknya Kedua Garismu Sejajar! Yahh, Persamaanmu Tidak Memiliki Solusi nih")
               
                x_16 = np.arange(c1_16-5,c1_16+5,0.01)
                plt.figure(figsize=(8,8),dpi=200)
                plt.plot(x_16, m1_16*(x_16 - tx1_16)+ty1_16, color = 'pink', label = 'y = {}x+{}'.format(m1_16,(m1_16*tx1_16)*-1 + ty1_16 ))
                plt.plot(x_16, m2_16*(x_16 - tx2_16)+ty2_16, color = 'blue', label = 'y = {}x+{}'.format(m2_16,(m2_16*tx2_16)*-1 + ty2_16 ))
                plt.axhline(0, color='black', linewidth=0.5)
                plt.axvline(0, color='black', linewidth=0.5)
                plt.grid(color='gray', linestyle='--', linewidth=0.5)
                plt.title('Grafik 2 Fungsi Linear', fontdict={'fontname':'Times New Roman', 'fontsize': 14}) #mengganti font yang digunakan pada judul grafik
                plt.xlabel('X')
                plt.ylabel('Y')
                plt.legend(loc='best')
                plt.show()
                
            else : #karena m1 =/= m2 sehingga memiliki solusi tunggal
 
                titikpotongx_16 = (c2_16-c1_16)/(m1_16-m2_16) #mencari titik potong x
                titikpotongy_16 = m1_16*(titikpotongx_16-tx1_16)+ty1_16   #mencari titik potong y
                print(f'Koordinat titik potong dari dua persamaan tersebut adalah ({titikpotongx_16}, {titikpotongy_16})') #menampilkan koorddinat titik potong

                x_16 = np.arange(titikpotongx_16-10,titikpotongx_16+10,0.01)
                plt.figure(figsize=(4,4),dpi=200)
                plt.plot(x_16, m1_16*(x_16 - tx1_16)+ty1_16, color = 'pink', label = 'y = {:.2f}x+{:.2f}'.format(m1_16, -m1_16*tx1_16+ty1_16))
                plt.plot(x_16, m2_16*(x_16 - tx2_16)+ty2_16, color = 'blue', label = 'y = {:.2f}x+{:.2f}'.format(m2_16, -m2_16*tx2_16+ty2_16))
                plt.plot(titikpotongx_16,titikpotongy_16, color='orange', marker='o',markersize=4) #memberi tanda pada titik potong
                plt.axhline(0, color='black', linewidth=0.5)
                plt.axvline(0, color='black', linewidth=0.5)
                plt.grid(color='gray', linestyle='--', linewidth=0.5)
                plt.title('Grafik 2 Fungsi Linear', fontdict={'fontname':'Times New Roman', 'fontsize': 14}) #mengganti font yang digunakan pada judul grafik
                plt.xlabel('X')
                plt.ylabel('Y')
                plt.legend(loc='best')
                plt.show()
        elif ask_16.lower() == "n" : #jika gradien bukan integer maka dikonversi menjadi bilangan bulat terlebih dahulu
            pembilang1_16 = int(input ("Masukkan nilai pembilang pada gradien pertama: "))
            penyebut1_16 = int(input("Masukkan nilai penyebut pada gradien pertama: "))
            m1_16 = pembilang1_16/penyebut1_16 #menghitung gradien jika berbentuk pecahan
            tx1_16 = float(input("Masukkan koordinat titik x pertama: "))
            ty1_16 = float(input('Masukkan koordinat titik y pertama: '))
            pembilang2_16 = int(input ("Masukkan nilai pembilang pada gradien kedua: "))
            penyebut2_16 = int(input("Masukkan nilai penyebut pada gradien kedua: "))
            m2_16 = pembilang2_16/penyebut2_16 #menghitung gradien jika berbentuk pecahan
            tx2_16 = float(input("Masukkan koordinat titik x kedua: "))
            ty2_16 = float(input('Masukkan koordinat titik y kedua: '))
            print("persamaan garis dengan gradien", m1_16, "dan pada koordinat titik (",{tx1_16},",",{ty1_16},") adalah y = {:.2f}x+{:.2f}".format(m1_16, -m1_16*tx1_16+ty1_16))
            print("persamaan garis dengan gradien", m2_16, "dan pada koordinat titik (",{tx2_16},",",{ty2_16},") adalah y = {:.2f}x+{:.2f}".format(m2_16, -m2_16*tx2_16+ty2_16))
            c1_16 = -m1_16*tx1_16 + ty1_16
            c2_16 = -m2_16*tx2_16 + ty2_16
            #mengevaluasi nilai gradien
            if m1_16 == m2_16 : 
                print ("Gradiennya sama nih! Jadi kemungkinan tidak ada solusi (sejajar) atau solusinya sangat banyak (berhimpit)!")
                if c1_16 == c2_16: #mencari tahu apakah kedua garis tersebut berhimpit atau sejajar dengan melihat nilai dari variabel c
                    print("Wahhh Kedua garis berhimpit! Jadi Persamaan Tersebut Memiliki Banyak Solusi")
                else : 
                    print("Tampaknya Kedua Garismu Sejajar! Yahh, Persamaanmu Tidak Memiliki Solusi nih")
                x_16 = np.arange(c1_16-5,c1_16+5,0.01)
                plt.figure(figsize=(8,8),dpi=200)
                plt.plot(x_16, m1_16*(x_16 - tx1_16)+ty1_16, color = 'pink', label = 'y = {}x+{}'.format(m1_16,(m1_16*tx1_16)*-1 + ty1_16 ))
                plt.plot(x_16, m2_16*(x_16 - tx2_16)+ty2_16, color = 'blue', label = 'y = {}x+{}'.format(m2_16,(m2_16*tx2_16)*-1 + ty2_16 ))
                plt.axhline(0, color='black', linewidth=0.5)
                plt.axvline(0, color='black', linewidth=0.5)
                plt.grid(color='gray', linestyle='--', linewidth=0.5)
                plt.title('Grafik 2 Fungsi Linear', fontdict={'fontname':'Times New Roman', 'fontsize': 14}) #mengganti font yang digunakan pada judul grafik
                plt.xlabel('X')
                plt.ylabel('Y')
                plt.legend(loc='best')
                plt.show()
            else : #karena m1 =/= m2 sehingga memiliki solusi tunggal
                titikpotongx_16 = (c2_16-c1_16)/(m1_16-m2_16) #mencari titik potong x
                titikpotongy_16 = m1_16*(titikpotongx_16-tx1_16)+ty1_16   #mencari titik potong y
                print(f'Koordinat titik potong dari dua persamaan tersebut adalah ({titikpotongx_16}, {titikpotongy_16})') #menampilkan koorddinat titik potong

                x_16 = np.arange(titikpotongx_16-5,titikpotongx_16+5,0.01)
                plt.figure(figsize=(4,4),dpi=200)
                plt.plot(x_16, m1_16*(x_16 - tx1_16)+ty1_16, color = 'pink', label = 'y = {:.2f}x+{:.2f}'.format(m1_16, -m1_16*tx1_16+ty1_16))
                plt.plot(x_16, m2_16*(x_16 - tx2_16)+ty2_16, color = 'blue', label = 'y = {:.2f}x+{:.2f}'.format(m2_16, -m2_16*tx2_16+ty2_16))
                plt.plot(titikpotongx_16,titikpotongy_16, color='orange', marker='o',markersize=4) #memberi tanda pada titik potong
                plt.axhline(0, color='black', linewidth=0.5)
                plt.axvline(0, color='black', linewidth=0.5)
                plt.grid(color='gray', linestyle='--', linewidth=0.5)
                plt.title('Grafik 2 Fungsi Linear', fontdict={'fontname':'Times New Roman', 'fontsize': 14}) #mengganti font yang digunakan pada judul grafik
                plt.xlabel('X')
                plt.ylabel('Y')
                plt.legend(loc='best')
                plt.show()
        else :
            print ("Masukkan anda salah! Terima Kasih!")
    elif pilihan_16 == "3":
        print(" Kondisi yang ingin kamu cari solusinya adalah mengikuti persamaan garis ax+by=c dan dx+ey=f")
        #mendefinisikan solusi dari persamaan yang akan dicari nantinya
        def solusi_16(a_16, b_16, c_16, d_16, e_16, f_16):
            # Membentuk matriks berdasarkan koefisien persamaan
            koe_16 = np.array([[a_16, b_16],
                               [d_16, e_16]])

            # Membentuk matriks berdasarkan konstanta persamaan
            kons_16 = np.array([c_16, f_16])

            # Menghitung determinan dari matriks koefisien yang kemudian dapat digunakan untuk mengevaluasi solusi persamaan
            detkoe_16 = np.linalg.det(koe_16) 

            if detkoe_16 != 0:
                # Solusi tunggal
                solusinya_16 = np.linalg.solve(koe_16, kons_16)
                hasil_16 = f"Persamaan {a_16}x+{b_16}y={c_16} dan {d_16}x+{e_16}y={f_16} memiliki solusi tunggal, yaitu ({solusinya_16[0]}, {solusinya_16[1]})"
                grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, solusinya_16) #memanggil fungsi grafik_hasil_16
                return hasil_16 #mengembalikan fungsi grafik_hasil_16
            else:
                # Cek apakah sistem konsisten
                cekpoin_16 = np.array([[a_16, b_16, c_16],
                                       [d_16, e_16, f_16]])
                rank_A_16 = np.linalg.matrix_rank(koe_16)
                rank_cekpoin_16 = np.linalg.matrix_rank(cekpoin_16)

                if rank_A_16 == rank_cekpoin_16:
                    hasil_16 = f"Persamaan {a_16}x+{b_16}y={c_16} dan {d_16}x+{e_16}y={f_16} memiliki solusi yang tidak terbatas "
                    grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, None) #memanggil fungsi grafik_hasil_16
                    return hasil_16 #mengembalikan fungsi grafik_hasil_16
                else:
                    hasil_16 = f"Persamaan {a_16}x+{b_16}y={c_16} dan {d_16}x+{e_16}y={f_16} tidak memiliki solusi"
                    grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, None) #memanggil fungsi grafik_hasil_16
                    return hasil_16 #mengembalikan fungsi grafik_hasil_16

        def grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, solusinya_16):
            # Menentukan rentang nilai x
            if solusinya_16 is not None:
                x_16 = np.linspace(solusinya_16[0] - 5, solusinya_16[0] + 5, 400)
            else:
                x_16 = np.linspace(-10, 10, 400)

            # Menghitung nilai y untuk setiap persamaan
            y1_16 = (c_16 - a_16 * x_16) / b_16
            y2_16 = (f_16 - d_16 * x_16) / e_16

            plt.figure(figsize=(8, 8))
            plt.plot(x_16, y1_16, label=f'{a_16}x + {b_16}y = {c_16}', color='green')
            plt.plot(x_16, y2_16, label=f'{d_16}x + {e_16}y = {f_16}', color='orange')

            if solusinya_16 is not None:
                plt.plot(solusinya_16[0], solusinya_16[1], 'bo') #'bo' untuk memberi warna biru pada solusinya yang berbentuk o
                plt.annotate(f'{solusinya_16[0]:.2f}, {solusinya_16[1]:.2f}', xy=(solusinya_16[0], solusinya_16[1]), xytext=(solusinya_16[0] + 1, solusinya_16[1]))

            plt.axhline(0, color='black', linewidth=0.5)
            plt.axvline(0, color='black', linewidth=0.5)
            plt.grid(color='gray', linestyle='--', linewidth=0.5)
            plt.legend()
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title('Grafik Persamaan Linear')
            plt.show()

        # Masukkan untuk persamaan pertama
        a_16 = float(input("Masukkan nilai koefisien dari x (a):"))
        b_16 = float(input("Masukkan nilai koefisien dari y (b):"))
        c_16 = float(input("Masukkan nilai konstanta (c) dari persamaan ax+by = c: "))
        # Masukkan untuk persamaan kedua
        d_16 = float(input("Masukkan nilai koefisien dari x (d):"))
        e_16 = float(input("Masukkan nilai koefisien dari y (e):"))
        f_16 = float(input("Masukkan nilai konstanta (f) dari persamaan dx+ey = f: "))

        jawaban_16 = solusi_16(a_16, b_16, c_16, d_16, e_16, f_16)
        print(jawaban_16)
    elif pilihan_16 == "4":
        print("Terima kasih dan sampai jumpa kembali!")
        mulai_16 = False
    else:
        print ("Maaf masukan anda salah!")
        
```

```python
#kasus 3
import numpy as np
import matplotlib.pyplot as plt

def solusi_16(a_16, b_16, c_16, d_16, e_16, f_16):
    # Membentuk matriks untuk koefisien persamaan
    koe_16 = np.array([[a_16, b_16],
                       [d_16, e_16]])
    
    # Membentuk matriks untuk konstanta persamaan
    kons_16 = np.array([c_16, f_16])
    
    # Menghitung determinan dari matriks koefisien yang kemudian dapat digunakan untuk mengevaluasi solusi persamaan
    detkoe_16 = np.linalg.det(koe_16)
    
    if detkoe_16 != 0:
        # Solusi tunggal
        solusinya_16 = np.linalg.solve(koe_16, kons_16)
        hasil_16 = f"Solusi tunggal: ({solusinya_16[0]}, {solusinya_16[1]})"
        grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, solusinya_16)
        return hasil_16
    else:
        # Cek apakah sistem konsisten
        cekpoin_16 = np.array([[a_16, b_16, c_16],
                               [d_16, e_16, f_16]])
        rank_A_16 = np.linalg.matrix_rank(koe_16)
        rank_cekpoin_16 = np.linalg.matrix_rank(cekpoin_16)

        if rank_A_16 == rank_cekpoin_16:
            hasil_16 = "Tak terbatas solusi"
            grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, None)
            return hasil_16
        else:
            hasil_16 = "Tidak ada solusi"
            grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, None)
            return hasil_16

def grafik_hasil_16(a_16, b_16, c_16, d_16, e_16, f_16, solusinya_16):
    # Menentukan rentang nilai x
    if solusinya_16 is not None:
        x_16 = np.linspace(solusinya_16[0] - 5, solusinya_16[0] + 5, 400)
    else:
        x_16 = np.linspace(-10, 10, 400)
    
    # Menghitung nilai y untuk setiap persamaan
    y1_16 = (c_16 - a_16 * x_16) / b_16
    y2_16 = (f_16 - d_16 * x_16) / e_16
    
    plt.figure(figsize=(8, 8))
    plt.plot(x_16, y1_16, label=f'{a_16}x + {b_16}y = {c_16}', color='green')
    plt.plot(x_16, y2_16, label=f'{d_16}x + {e_16}y = {f_16}', color='orange')
    
    if solusinya_16 is not None:
        plt.plot(solusinya_16[0], solusinya_16[1], 'bo')
        plt.annotate(f'{solusinya_16[0]:.2f}, {solusinya_16[1]:.2f}', xy=(solusinya_16[0], solusinya_16[1]), xytext=(solusinya_16[0] + 1, solusinya_16[1]))
    
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Grafik Persamaan Linear')
    plt.show()

# Masukkan untuk persamaan pertama
a_16 = float(input("Masukkan nilai koefisien dari x (a):"))
b_16 = float(input("Masukkan nilai koefisien dari y (b):"))
c_16 = float(input("Masukkan nilai konstanta (c) dari persamaan ax+by = c: "))
# Masukkan untuk persamaan kedua
d_16 = float(input("Masukkan nilai koefisien dari x (d):"))
e_16 = float(input("Masukkan nilai koefisien dari y (e):"))
f_16 = float(input("Masukkan nilai konstanta (f) dari persamaan dx+ey = f: "))

jawaban_16 = solusi_16(a_16, b_16, c_16, d_16, e_16, f_16)
print(jawaban_16)
```
